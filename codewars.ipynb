Задача про определение ранга для новых членов клуба крокета.
Первым идёт мой вариант программы. Мне не нравится, что у меня выводится только один вариант.
Следом идет скрин с сайта, где видно что программу приняли в обработку и она ок.
Затем вариан программы самый залайканный. В нем выводятся весь список. 
И еще интересно как написан return, а именно ...for (age, handicup) in data. 
Так написано словно программа сама знает, раз это пара, то певое значение сравниать 
с первым сравнителем, а второе со вторым.

```python
def open_or_senior(data):
    output = []
    for item in data:
        if item[0] >= 55 and item[1] > 7:
            output.append('Senior')
        else:
            output.append('Open')
        return output
    
print(open_or_senior([(55, 21), (45, 12), (19, -2), (104, 20)]))
```

```python
def openOrSenior(data):
    return ["Senior" if age >= 55 and handicap >= 8 else "Open" for (age, handicap) in data]

print(openOrSenior([(45, 12), (55, 21), (19, -2), (104, 20)]))
```
